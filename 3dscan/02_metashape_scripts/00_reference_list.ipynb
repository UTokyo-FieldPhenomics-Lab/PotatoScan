```python
import os
import pandas as pd
import numpy as np
```

```python
working_directory = r"/home/crest/z/hwang_Pro/data/2023_hokkaido_potato"
img_folder = os.path.join(working_directory, 'images')
mask_folder = os.path.join(working_directory, 'masks')
```

```python
foldernames = os.listdir(img_folder)
foldernames
```

```python
sb = pd.DataFrame(
    columns=['id', 'scalebar', 'ff', 'front', 'back']).astype(
        {'id': 'str', 'scalebar': 'str', 'ff':'uint8', 'front': 'uint8', 'back': 'uint8'})
for i in foldernames:
    front, back = i.split('-')
    ff, front = front.split('R')
    if ff == '':
        ff = 1

    ff, front, back = int(ff), int(front), int(back)
        
    sb.loc[len(sb)] = [i, 'scalebar.csv', ff, front, back]

sb2 = sb.sort_values(['ff', 'front', 'back'], ascending=[True, True, True])
```

```python
for foldername, subfolders, filenames in os.walk(img_folder):
    for filename in filenames:
        file_path = os.path.join(foldername, filename)

        if '.JPG' in file_path:
            new_name = file_path.replace('.JPG', '.jpeg')
            os.rename(file_path, new_name)
            print(new_name)
```

```python
for foldername, subfolders, filenames in os.walk(img_folder):
    for filename in filenames:
        file_path = os.path.join(foldername, filename)

        if '.jpeg' in file_path:
            new_name = file_path.replace('.jpeg', '.jpg')
            os.rename(file_path, new_name)
            print(new_name)
```

```python
for foldername, subfolders, filenames in os.walk(mask_folder):
    for filename in filenames:
        file_path = os.path.join(foldername, filename)

        if '.JPG' in file_path:
            new_name = file_path.replace('.JPG', '.png')
            os.rename(file_path, new_name)
            print(new_name)
```
