```python
import Metashape
import pandas as pd
import numpy as np

import configs as cfg
import os
```

```python
ms_list = ['1R_Group0', '1R_Group1', '1R_Group2', 
           '2R_Group0',
           '3R_Group0', '3R_Group1',
           '4R_Group0', '4R_Group1',
           '5R_Group0', 
           'XR_GroupFix']
```

```python
vol_dict = {}

for ms_name in ms_list:
    doc = Metashape.Document()
    doc.open(os.path.join(cfg.working_directory, "projects.psx", ms_name + '.psx'))

    for chunk in doc.chunks:
        vol_dict[chunk.label] = chunk.model.volume()
```

```python
vol_dict['R1-1'], type(vol_dict['R1-1'])
```

read scalebarlist.csv to get the time-series order

```python
sbl = pd.read_csv('scalebarlist.csv', header=None)
sbl
```

```python
volume_csv = pd.DataFrame(columns=[['label', 'group_id', 'bucket_id', 'pin_id', 'volume (cm3)']])

for i, label in enumerate(sbl[0]):
    vol_cm = vol_dict[label] * 1000 * 1000

    if vol_cm == 0:
        print(label)

    volume_csv.loc[len(volume_csv),:] = [label, sbl.loc[i, 2], sbl.loc[i, 3], sbl.loc[i, 4], vol_cm]
```

```python
volume_csv.to_csv(os.path.join(cfg.working_directory, "outputs", "volume.csv"), index=None)
```
